# Transforming Data
# 1. Removing Special Characters from Data 

```python
df["discounted_price"]=df["discounted_price"].str.replace("₹" , "")
```

```python
df["actual_price"]=df["actual_price"].str.replace("₹" , "")
```

```python
df["rating"]=df["rating"].str.replace("." , "")
```

```python
df['discounted_price']
```

```python
df['actual_price']
```

```python
df['rating'].unique()
```

```python
# Check the strange row
df.query('rating == "|"')
```

```python
# replace with the actual value from the source website
# the actual value from the source is 3.9 , we put it with 39 to convert it to float and divided by 10 
# ,so we get the actual value we need
df["rating"]=df["rating"].str.replace("|" , "39")
```

```python
df.head(5)
```

# 2. Changing Data Types of Columns

```python
df["rating"]=df["rating"].astype(float)
df["rating"]=df["rating"]/10
```

```python
df["rating"].unique()
```

```python
df["discounted_price"]=df["discounted_price"].str.replace("," , "")
df["discounted_price"]=df["discounted_price"].astype(float)
```

```python
df["discounted_price"]
```

```python
df["actual_price"]=df["actual_price"].str.replace("," , "")
df["actual_price"]=df["actual_price"].astype(float)
```

```python
df["actual_price"]
```

```python
df["rating_count"]=df["rating_count"].str.replace("," , "")
df["rating_count"]=df["rating_count"].astype(float)
```

```python
df["rating_count"]
```

```python
df["discount_percentage"]=df["discount_percentage"].str.replace("%" , "")
df["discount_percentage"]=df["discount_percentage"].astype(float)
df["discount_percentage"]=df["discount_percentage"]/100
```

```python
df["discount_percentage"]
```

```python
df.head(5)
```

# 3. Splitting the Category Column into Category and Subcategory

```python
df['category']
```

```python
df[['Category','Sub Category']] = df['category'].str.split("|",expand=True).loc[:,:1]
```

```python
df.head(5)
```

# 4. Dropping Unnecessary Columns

```python
df_original = df
df=df.drop('category',axis=1)
df=df.drop('about_product',axis=1)
df=df.drop('review_id',axis=1)
df=df.drop('review_content',axis=1)
df=df.drop('review_title',axis=1)
df=df.drop('user_name',axis=1)
```

```python
df.head()
```

```python
df[df['rating_count'].isnull()]
```
